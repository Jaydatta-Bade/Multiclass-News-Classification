## Classification of News Articles
Multiclass Classification of News articles which are having classes numbered 1 to 4, where
- 1 is "World News"
- 2 is "Sports News"
- 3 is "Business News"
- 4 is "Science-Technology News"

The classification algorithms used are:-
- Multinomial Naive Bayes 
- Decision Tree 
- Gaussian Naive Bayes 
- Stochastic Gradient Descent Classifier 
- LGBM (light gradient boosting machine) Classifier

### WordCloud of News Articles of Different Types

```python
stopwordset= set(STOPWORDS)
morestop={'lt','gt','href','HREF','quot','aspx'}
stopwordset= stopwordset.union(morestop)
```

```python
# world news
world = df.description[df.classid[df.classid==1].index]
```

```python
plt.figure(figsize = (12,15)) ;
wordcloud = WordCloud(min_font_size = 3,  max_words = 500 ,background_color="white",contour_width=3, contour_color='firebrick', stopwords=stopwordset,colormap='Dark2').generate(" ".join(world))
plt.imshow(wordcloud,interpolation = 'bilinear');
plt.axis("off");
```

```python
# sports news
sports = df.description[df.classid[df.classid==2].index]
```

```python
plt.figure(figsize = (12,15)) ;
wordcloud = WordCloud(min_font_size = 3,  max_words = 500 ,background_color="white",contour_width=3, contour_color='skyblue', stopwords=stopwordset,colormap='Dark2').generate(" ".join(sports))
plt.imshow(wordcloud,interpolation = 'bilinear');
plt.axis("off");
```
